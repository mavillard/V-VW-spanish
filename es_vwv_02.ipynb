```python
import pickle
import string

import nltk
from nltk.corpus import stopwords
```

```python
with open('data/out/stories.pickle', 'rb') as f:
    stories = pickle.load(f)
```

```python
len(stories)
```

```python
stop = stopwords.words('spanish') + list(string.punctuation)

def clean(s):
    aux = s.lower()
    words = [w for w in nltk.word_tokenize(aux) if w not in stop]
    aux = ' '.join(words)
    r = aux.replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u')
    return r

def join_texts(d):
    return ' '.join(d.values())
```
